```python
from bs4 import BeautifulSoup

# Abre el archivo HTML
with open('guion.html', 'r', encoding='utf-8') as archivo:
    contenido_html = archivo.read()

# Crea un objeto BeautifulSoup
soup = BeautifulSoup(contenido_html, 'html.parser')

# Extrae el texto del HTML
texto_plano = soup.get_text()

# Define el nombre del archivo de salida
nombre_archivo_salida = 'texto_plano.txt'

# Guarda el texto plano en un archivo .txt
with open(nombre_archivo_salida, 'w', encoding='utf-8') as archivo_salida:
    archivo_salida.write(texto_plano)

print(f"Texto plano guardado en {nombre_archivo_salida}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Lee el archivo CSV
data = pd.read_csv('estadisticas-nba.csv')

# Extrae las columnas Player y PTS (puntos)
jugadores = data['Player']
puntos = data['PTS']

# Limita la cantidad de jugadores a mostrar en la gráfica (por ejemplo, los 20 mejores)
top_jugadores = 20
jugadores = jugadores[:top_jugadores]
puntos = puntos[:top_jugadores]

# Crea la gráfica de barras
plt.figure(figsize=(12, 6))
plt.barh(jugadores, puntos, color='skyblue')
plt.xlabel('Puntos')
plt.ylabel('Jugadores')
plt.title('Puntajes de Jugadores de Baloncesto')
plt.gca().invert_yaxis()  # Invierte el eje y para mostrar al mejor jugador en la parte superior
plt.tight_layout()

# Muestra la gráfica
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Lee el archivo CSV
data = pd.read_csv('NFL-STAT.csv')

# Extrae las columnas Player_id y Pass_long (puntos)
jugadores = data['player_id']
puntos = data['pass_long']

# Limita la cantidad de jugadores a mostrar en la gráfica (por ejemplo, los 20 mejores)
top_jugadores = 20
jugadores = jugadores[:top_jugadores]
puntos = puntos[:top_jugadores]

# Crea la gráfica de barras
plt.figure(figsize=(12, 6))
plt.barh(jugadores, puntos, color='skyblue')
plt.xlabel('Puntos')
plt.ylabel('Jugadores')
plt.title('Pases de los jugadores')
plt.gca().invert_yaxis()  # Invierte el eje y para mostrar a los jugadores con mas pases
plt.tight_layout()

# Muestra la gráfica
plt.show()
```
